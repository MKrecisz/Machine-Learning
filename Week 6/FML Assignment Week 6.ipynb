_Naïve Bayes_ tries to pick the words that are related to the given lables. In this example it will try to use bags-of-words that base on frequency and connect it's relation with one of two given groups (positive and negative ratings).

```python
import sklearn as sk
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns
```

For this ML we need to filter the df just to **Dresses** and divide between **Ratings**. Since we will be using _Naïve Bayes_ we will need only **Review Text** column and the **Rating** one. 

```python
df = pd.read_csv('Assignment text mining - data clothing reviews.csv')
df = df[df['Class Name'].str.match('Dresses', na=False)]
df['Rating'] = df['Rating'].replace([1,2,3], 0)
df['Rating'] = df['Rating'].replace([4,5], 1)

df = df[['Rating', 'Review Text']]
df = df.dropna()
df.head()
```

```python
text = df['Review Text'].values.astype('U') 
vect = CountVectorizer(stop_words='english') 
vect = vect.fit(text) 

docu_feat = vect.transform(text)
print(docu_feat[0:50,0:50])
```

```python
docu_feat = pd.DataFrame(docu_feat.toarray())
```

```python
rev_words = pd.concat([df, docu_feat], axis=1)
rev_wrods= rev_words.dropna()
rev_words.head(10)
#I was wondering where from all those NaNs are coming from
```

```python
rev_words = rev_words.dropna()
rev_words
```

## Model

```python
nb = MultinomialNB()
X = rev_words.iloc[:,1:] 
y = rev_words['Rating']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20) #split the data and store it


final = pd.DataFrame()
final['Review Text'] = X_test['Review Text']
X_train = X_train.drop(['Review Text'], axis = 1)
X_test = X_test.drop(['Review Text'], axis = 1)


nb = nb.fit(X_train, y_train)
y.shape, X.shape
```

```python
y_test_p = nb.predict(X_test)
nb.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_test_p)
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
print(classification_report(y_test, y_test_p))
```

```python
final['predicted'] = y_test_p
final['true_y'] = y_test
final.reset_index(drop=True)
final = final.loc[~(final['predicted'] == final['true_y'])]
final = final.reset_index(drop=True)
final
```
