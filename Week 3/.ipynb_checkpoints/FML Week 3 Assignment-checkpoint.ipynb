```python
#importing everything that's necessary
import seaborn as sns 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import math

#loading 1st df
df = pd.read_csv('data Mashable.csv')
df.head()
```

```python
dflength = len(df.index)
dftraininglength = round(len(df.index)*0.8)
dftestlength = len(df.index)-round(len(df.index)*0.8)

dftraining = df.iloc[:dftraininglength]
dftest = df.iloc[dftraininglength:dflength]
print("The length of the rows in this dataframe is: ")
print(dflength)
print("The training set will be about 80 percent of it which is: ")
print(dftraininglength)
print("The testing set will be the rest which is:")
print(dftestlength)
```

```python
dfdummies = df[["data_channel_is_lifestyle", "data_channel_is_entertainment", "data_channel_is_bus", "data_channel_is_socmed", "data_channel_is_tech"]]
dffinal = pd.concat([df["shares"], dfdummies], axis = "columns")
dffinal.head()
```

```python
#To use sk-learn, we need to create lists of the two variables
#Note that you need to use [[ ]] with x. This creates a Pandas Data Frame (table), rather than a Pandas Series (list).
#lm.fit needs a Data Frame (or matrix) for x, because x can be multiple variables
x = df[["data_channel_is_lifestyle", "data_channel_is_entertainment", "data_channel_is_bus", "data_channel_is_socmed", "data_channel_is_tech"]]
y = df["shares"]
lm = LinearRegression() #this creates a new LR model
lm.fit(x, y) #this "fits" the model

b0 = lm.intercept_ #gets the intercept
b1 = lm.coef_[0] #lm.coef_ gets a list of coefficients (excluding the intercept). [0] gets the actual number from the list

print(f"The regression line is equal to y = {b0:.3f} + {b1:.3f}X") #.0f formats to 3 decimals. 
```

```python
print("Predicted shares for a lifestyle channel:")
print(lm.predict([[1,0,0,0,0]]))
print("r2 score is:")
print(lm.score(x, y))
```

```python
df['shares_p'] = lm.predict(x) #generate the predictions and add them to the dataframe
df['resid'] = df['shares'] - df['shares_p'] #get the residuals by subtracting the prediction from the actual views
print("r2 score obtained through a different method:")
print(r2_score(df['shares'], df['shares_p'])) #this gives the model fit (R2) for the Y and predicted Y
```

```python
#RMSE
math.sqrt(mean_squared_error(df['shares'], df['shares_p']))
```

```python
dummydummy = pd.get_dummies(df[["data_channel_is_lifestyle", "data_channel_is_entertainment"]])
dummydummy.head(10)
dffinal = pd.concat([df["shares"], dummydummy], axis = "columns")
dffinal.head()
```

```python
#To use sk-learn, we need to create lists of the two variables
#Note that you need to use [[ ]] with x. This creates a Pandas Data Frame (table), rather than a Pandas Series (list).
#lm.fit needs a Data Frame (or matrix) for x, because x can be multiple variables
x = dffinal[["data_channel_is_lifestyle", "data_channel_is_entertainment"]]
y = dffinal["shares"]
lm = LinearRegression() #this creates a new LR model
lm.fit(x, y) #this "fits" the model

b0 = lm.intercept_ #gets the intercept
b1 = lm.coef_[0] #lm.coef_ gets a list of coefficients (excluding the intercept). [0] gets the actual number from the list

print(f"The regression line is equal to y = {b0:.3f} + {b1:.3f}X") #.0f formats to 3 decimals. 
```

```python
print("Predicted shares for a lifestyle channel:")
print(lm.predict([[1,0]]))
print("r2 score is:")
print(lm.score(x, y))
```
