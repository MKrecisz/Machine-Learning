```python
import seaborn as sns #this is the plotting library I'll be using 
import pandas as pd #"as pd" means that we can use the abbreviation in commands
import matplotlib.pyplot as plt #we need Matplotlib for setting the labels in the Seaborn graphs

#loading 1st df
df1 = pd.read_csv('steps.csv', sep=";")
dfavg = df1.drop("id", axis = 1).mean(axis = 1)
df1["average"]=dfavg
df1.head()
```

```python
#loading 2nd df
df2 = pd.read_csv('survey.csv')
df2.head()
```

```python
#merging dfs
df = pd.merge(df1, df2, on = 'id') #'id' is the common identifier of two dfs
df.head()
```

```python
df.info() #Get information on the variables
```

```python
df['average'].value_counts() #Get all the values. This may be less useful when there are many values.
```

```python
df['average'] = df['average'].apply(pd.to_numeric, errors='coerce')
dfavg = df['average']

def above_25000(x):
    if(x > 25000): 
        return float('NaN')
    else: 
        return x
dfavg = dfavg.apply(above_25000)
mean_size = dfavg.mean()
median_size = dfavg.median()
avgvar = dfavg.var()
avgstd = dfavg.std()
print(f'The median is: {median_size}')
print(f'The mean is: {mean_size}')
print(f'The variance of average steps per user is: {avgvar}')
print(f'Standard diviation is: {avgstd}')
```

```python
#sns.histplot(df['average'].dropna(), kde=False) #Selecting the rating column. I need to drop the NA's for the plot
sns.distplot(dfavg, kde=False, vertical = False)
plt.title('Number of people per average amount of steps')
plt.xlabel('Number of steps')
plt.ylabel('Number of people')
plt.axvline(median_size, 0, 100, color='green', label='median') #This adds a vertical line at x-position median_size, from y = 0 to y = 100 
plt.axvline(mean_size, 0, 100, color='blue', label='mean') 
plt.legend() #This adds a legend. It works automatically because I've set the labels in the previous lines
plt.show()



```

```python
sns.boxplot(dfavg)
plt.show()
```

We can see in those two graphs that the data makes it right skewed, so that means that the majority of people who contributed to gathering this data make averagely more than 6240.073033707865 steps daily.
