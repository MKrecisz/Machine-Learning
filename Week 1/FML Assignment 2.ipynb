```python
import seaborn as sns #importing everything that's necessary
import pandas as pd 
import matplotlib.pyplot as plt

#loading 1st df
df1 = pd.read_csv('steps.csv', sep=";")
#alreqady preparing average steps
dfavg = df1.drop("id", axis = 1).mean(axis = 1)
df1["average"]=dfavg
#loading 2nd df
df2 = pd.read_csv('survey.csv')
#merging dfs
df = pd.merge(df1, df2, on = 'id') #'id' is the common identifier of two dfs
df.head()
```

```python
#cleaning data
def above_100000(x):
    if(x > 100000): 
        return float('NaN')
    else: 
        return x
    
def above_theheaviestmanonearth(x):
    if(x > 442): 
        return float('NaN')
    else: 
        return x
    
df['average'] = df['average'].apply(above_100000)
df['weight'] = df['weight'].apply(above_theheaviestmanonearth)
```

```python
#pairplot just for a better outline of things
sns.pairplot(df[["average", "weight", "height" ]])
```

```python
#correlation matrix
corr = df[["average", "weight", "height" ]].corr()
corr
```

```python
sns.scatterplot(data=df, x="average", y="weight")
```

```python
sns.scatterplot(data=df, x="average", y="height")
```

```python
dfselected = df[["average", "weight", "height" ]]
dfselected = dfselected.dropna(axis=0)
from scipy.stats.stats import pearsonr
print("Pearson's r for average steps and weight: ")
print(pearsonr(dfselected["average"], dfselected["weight"]))

print("Pearson's r for average steps and height: ")
print(pearsonr(dfselected["average"], dfselected["height"]))
```

Pearson's r indicate that in both examples the correlation is not very significant. The strength of the association despite being stronger between average steps and height is also not very promising. Regarding the linearity, the first plot is not very linear and the second - showing a better distribution on the y axis - is.
